## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on = "ISBN")
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df = num_rating_df.rename(columns={'Book-Rating' : 'num_ratings'})
```

```python
avg_rating_df = ratings_with_name.groupby("Book-Title")['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns={'Book-Rating' : 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="Book-Title")
popular_df
```

```python
#Here we are taking the most popular books (with no.of ratings>250 sorted by avg_rating desc and taking top 50 in the list)
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on = "Book-Title").drop_duplicates("Book-Title")[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df
```

## Collaborative Filtering Based Recommender System

```python
#Getting users who have given more than 200 ratings
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
#getting Books that have got at least 50 ratings

y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Here each book is like represented in 810 dimesions
#706-books and 810 columns
#Each book is a vector in a 810 dimesion
#we have to find the similarity between each vector(Book) with other vectors(book)(each row with other rows)
#Like distance between them is the similarity between them

similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
#Recommend Function

def recommend(book_name):
  #index_fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_score[index])), key= lambda x:x[1], reverse=True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```

```python
x = recommend('The Da Vinci Code')
```
